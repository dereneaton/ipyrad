## Demultiplex on i7 outer barcodes 

Outer barcodes/indicies (e.g., `i7` barcodes) can be used to label samples for demultiplexing, or, often to label different libraries. An example could be you and a friend both prepared separate libraries for different organisms but used the same set of 48 internal barcodes. You then both add a unique outer barcode to your samples so that you can pool them together for sequencing.
